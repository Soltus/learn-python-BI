## 3.数据整理

整合两张movie表

```python
movie = pd.merge(movie1, movie2, how='inner', left_on='NAME', right_on='电影名')  #inner取两张表的键的交集
movie
```

```python
movie.drop(['ADD_TIME','NAME'],axis=1,inplace=True)
movie
```

```python
movie.sort_values(by='ID',inplace=True)
movie
```

```python
movie.reset_index(drop=True,inplace=True)   #重建索引
movie
```

```python
movie['重复'] = 0   
t = 0
for i in range(movie.shape[0]-1):   #movie.shape[0]为行数
    if t==0:
        k = i     #i为循环次数的变量，每循环一次就会增加1，为了让此时的i的值固定住，将其赋给k
        t = 1
        
    if movie.iloc[k,0] == movie.iloc[i+1,0]:
        movie.iloc[k,1] = movie.iloc[k,1]+'|'+movie.iloc[i+1,1]   #添加类型
        movie.iloc[k,5] = movie.iloc[k,5]+'|'+movie.iloc[i+1,5]   #添加特色
        movie.iloc[i+1,8] = 1    #将i+1这行的movie['重复']=1
    else:
        t = 0
```

```python
movie.head(20)
```

```python
movie0 = movie[movie['重复']==0]
movie0
```

```python
movie0.reset_index(drop=True,inplace=True)   #重建索引
movie0
```

```python
movie0.drop('重复',axis=1,inplace=True)
movie0
```

```python
for i in range(movie0.shape[0]):
    movie0.iloc[i,1] = '|'.join(set(movie0.iloc[i,1].split('|')))
    movie0.iloc[i,5] = '|'.join(set(movie0.iloc[i,5].split('|')))
```

```python
movie0
```

求comment中RATING的标准差

```python
comment.groupby('MOVIEID')['RATING'].std()
```

```python
comment.columns
```

```python
comment['RATING'] = pd.to_numeric(comment['RATING'])
```

```python
comment1 = comment[comment['RATING']!='-']
```

```python
comment1['RATING'] = pd.to_numeric(comment1['RATING'])
```

```python
rating = comment1.groupby('MOVIEID')['RATING'].std()
```

```python
rating
```

```python
rating = comment1.groupby('MOVIEID')['RATING'].std().to_frame('评分标准差').reset_index()
```

将“评分标准差”添加入movie0

```python
movie_new = pd.merge(movie0, rating, how='inner', left_on='ID', right_on='MOVIEID')
movie_new
```

```python
movie_new[['ID','电影名','地区','导演','主演','类型','特色','评分','评分标准差']].to_csv('./data/movie_new.csv',index=False,encoding='utf-8')
```
