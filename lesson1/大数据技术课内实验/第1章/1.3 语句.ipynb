```python
import pypinyin  #调用第三方包
a = ['胡志和','李天瑜','石宏伟','陈九昌','陈九雪','胡志超','李媛','郭天营','黄显玉','尹俊','左太强','陈映芳','罗坤','张正蕊','李福忠','张文升','严世翔','鲁升升','户志和','李华慧','姜世盈']
b = []
```

```python
for item in a:  #本循环完成从中文字符到拼音的转换
    s = ''
    for i in pypinyin.pinyin(item): #将单个汉字链接为姓名
        s = s+ ''.join(i)
    b.append(s)
print(a)
print(b)
```

```python
#冒泡排序    
for right in range(len(b),0,-1):  #外循环：确定排序轮数和内循环次数
    for j in range(0,right-1): #内循环：通过两两交换将最大值放在最后
        if b[j] > b[j+1]:
            b[j],b[j+1] = b[j+1],b[j]   #声调大小：轻三一二四
            a[j],a[j+1] = a[j+1],a[j]   #对应中文列表的调整
            
print(a)
```
