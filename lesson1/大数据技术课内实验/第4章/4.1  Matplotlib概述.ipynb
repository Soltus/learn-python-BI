```python
import numpy as np
import matplotlib.pyplot as plt
data = np.arange(0,1.1,0.01)

plt.title('lines') ## 添加标题
plt.xlabel('x')## 添加x轴的名称
plt.ylabel('y')## 添加y轴的名称
plt.plot(data,data**2)## 添加y=x^2曲线
plt.plot(data,data**4)## 添加y=x^4曲线
plt.legend(['y=x^2','y=x^4'])#添加图例

plt.savefig('./data/y=x^2.png')  #保存图形到文件
plt.show()
```

```python
rad = np.arange(0,np.pi*2,0.01)
##第一幅子图
p1 = plt.figure(figsize=(4,5),dpi=80)## 确定画布大小
ax1 = p1.add_subplot(2,1,1)## 创建一个两行1列的子图，并开始绘制第一幅
plt.title('lines')## 添加标题
plt.xlabel('x')## 添加x轴的名称
plt.ylabel('y')## 添加y轴的名称
plt.plot(rad,rad**2)## 添加y=x^2曲线
plt.plot(rad,rad**4)## 添加y=x^4曲线
plt.legend(['y=x^2','y=x^4'])

##第二幅子图
ax2 = p1.add_subplot(2,1,2)## 创开始绘制第2幅
p1.subplots_adjust(hspace = 0.5)#设定两幅子图的间距
plt.title('sin/cos') ## 添加标题
plt.xlabel('rad')## 添加x轴的名称
plt.ylabel('value')## 添加y轴的名称
plt.plot(rad,np.sin(rad))## 添加sin曲线
plt.plot(rad,np.cos(rad))## 添加cos曲线
plt.legend(['sin','cos'])

plt.show()
```

```python
## 原图
x = np.linspace(0, 4*np.pi)## 生成x轴数据
y = np.sin(x)## 生成y轴数据
plt.plot(x,y,label="$sin(x)$")## 绘制sin曲线图
plt.title('sin')
plt.savefig('./data/默认sin曲线.png')
plt.show()
```

```python
## 修改rc参数后的图
plt.rcParams['lines.linestyle'] = '-.'
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['lines.marker'] = 's' #正方形
plt.plot(x,y,label="$sin(x)$",color='#FF00FF')## 绘制三角函数
plt.title('sin')
plt.savefig('./data/修改rc参数后sin曲线.png')
plt.show()
```

```python
##设置rc参数显示中文标题
## 设置字体为SimHei显示中文
plt.rcParams['font.size'] = 16
plt.rcParams['font.sans-serif']='SimHei' #设置黑体
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号
plt.plot(x,y,label="$sin(x)$",color='#FF00FF') # 绘制三角函数
plt.title('sin曲线',color = 'blue')
plt.savefig('./data/显示中文标题sin曲线.png')
plt.show()
```

```python
##设置rc参数显示中文标题
## 设置字体为SimHei显示中文
from matplotlib import font_manager
stfont = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\simsun.ttc',size=16) #设置宋体
plt.rcParams['axes.unicode_minus'] = False    # 正常显示负号
plt.plot(x,y,label="$sin(x)$",color='#FF00FF') # 绘制三角函数
plt.title('sin曲线',font= stfont)
plt.xlabel('x变量',font = stfont)
plt.ylabel('y变量',font = stfont)
plt.legend(['正弦'],prop = stfont) 
plt.savefig('./data/显示中文标题sin曲线.png')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,5,10,17,26]
plt.plot(x,y) 
for a,b in zip(x,y):  #显示数值
    plt.text(a+0.1,b,b,fontsize=12)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,5,10,17,26]
plt.plot(x,y)
plt.grid(axis = 'y', linestyle = '-.')
for a,b in zip(x,y):  #显示数值
    plt.text(a+0.1,b,b,fontsize=12)
plt.show()
```
