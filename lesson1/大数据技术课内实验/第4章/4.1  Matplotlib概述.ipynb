```python
import numpy as np
import matplotlib.pyplot as plt
data = np.arange(0,1.1,0.01)

plt.title('lines') ## 
plt.xlabel('x')## x
plt.ylabel('y')## y
plt.plot(data,data**2)## y=x^2
plt.plot(data,data**4)## y=x^4
plt.legend(['y=x^2','y=x^4'])#

plt.savefig('./data/y=x^2.png')  #
plt.show()
```

```python
rad = np.arange(0,np.pi*2,0.01)
##
p1 = plt.figure(figsize=(4,5),dpi=80)## 
ax1 = p1.add_subplot(2,1,1)## 1
plt.title('lines')## 
plt.xlabel('x')## x
plt.ylabel('y')## y
plt.plot(rad,rad**2)## y=x^2
plt.plot(rad,rad**4)## y=x^4
plt.legend(['y=x^2','y=x^4'])

##
ax2 = p1.add_subplot(2,1,2)## 2
p1.subplots_adjust(hspace = 0.5)#
plt.title('sin/cos') ## 
plt.xlabel('rad')## x
plt.ylabel('value')## y
plt.plot(rad,np.sin(rad))## sin
plt.plot(rad,np.cos(rad))## cos
plt.legend(['sin','cos'])

plt.show()
```

```python
## 
x = np.linspace(0, 4*np.pi)## x
y = np.sin(x)## y
plt.plot(x,y,label="$sin(x)$")## sin
plt.title('sin')
plt.savefig('./data/sin.png')
plt.show()
```

```python
## rc
plt.rcParams['lines.linestyle'] = '-.'
plt.rcParams['lines.linewidth'] = 3
plt.rcParams['lines.marker'] = 's' #
plt.plot(x,y,label="$sin(x)$",color='#FF00FF')## 
plt.title('sin')
plt.savefig('./data/rcsin.png')
plt.show()
```

```python
##rc
## SimHei
plt.rcParams['font.size'] = 16
plt.rcParams['font.sans-serif']='SimHei' #
plt.rcParams['axes.unicode_minus'] = False    # 
plt.plot(x,y,label="$sin(x)$",color='#FF00FF') # 
plt.title('sin',color = 'blue')
plt.savefig('./data/sin.png')
plt.show()
```

```python
##rc
## SimHei
from matplotlib import font_manager
stfont = font_manager.FontProperties(fname='C:\\Windows\\Fonts\\simsun.ttc',size=16) #
plt.rcParams['axes.unicode_minus'] = False    # 
plt.plot(x,y,label="$sin(x)$",color='#FF00FF') # 
plt.title('sin',font= stfont)
plt.xlabel('x',font = stfont)
plt.ylabel('y',font = stfont)
plt.legend([''],prop = stfont) 
plt.savefig('./data/sin.png')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,5,10,17,26]
plt.plot(x,y) 
for a,b in zip(x,y):  #
    plt.text(a+0.1,b,b,fontsize=12)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,5,10,17,26]
plt.plot(x,y)
plt.grid(axis = 'y', linestyle = '-.')
for a,b in zip(x,y):  #
    plt.text(a+0.1,b,b,fontsize=12)
plt.show()
```
