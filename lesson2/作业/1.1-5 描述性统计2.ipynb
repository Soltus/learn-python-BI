```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
data = pd.read_csv('./data/1.1/data2.csv')
data.head(5)
```

```python
#店铺平均销量
data_tmp = data[['店铺名','品类','半年销量']]
data_tmp1 = data_tmp.groupby(by = ['品类','店铺名'],as_index = False).mean()
```

```python
data_tmp1.groupby(by = ['品类'],as_index = False).mean()
```

```python
data_tmp1 = data_tmp[data_tmp['品类']=='水产'].sort_values(by=['半年销量'],ascending = False)
```

```python
data_tmp2 = data_tmp1.head(20)
data_tmp2.reset_index(drop=True,inplace = True)
```

```python
s = data_tmp[data_tmp['品类']=='水产']['半年销量'].sum()
data_tmp2['累计占比'] = 0.0
k = 0
for i in range(len(data_tmp2)):
    k = k + data_tmp2['半年销量'][i]
    data_tmp2['累计占比'][i] = k/s
```

```python
data_tmp2[['店铺名','累计占比']]
```
