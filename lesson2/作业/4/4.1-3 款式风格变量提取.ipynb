```python
import pandas as pd
```

```python
data = pd.read_csv('./data/4.1/data.csv')
```

```python
wordNum = pd.read_csv('./data/4.1/wordNum.csv')
```

```python
data
```

```python
styleList = []
for i in range(42):
    styleList.append(wordNum['关键词'][i])
```

```python
styleList
```

```python
styleList.remove('聚酯纤维')
styleList.remove('聚氨酯')
styleList.remove('氨纶')
styleList.remove('粘胶纤维')
styleList.remove('粘纤')
styleList.remove('纤维')
```

```python
print(styleList)
```

```python
for i in range(len(styleList)):
    data[styleList[i]] = 0
```

```python
data
```

```python
for i in range(len(data)):
    for j in range(len(styleList)):
        if (data['goodsName'][i].find(styleList[j]) != -1) or (data['productDescription'][i].find(styleList[j]) != -1):
            data[styleList[j]][i] = 1
```

```python
data
```

```python
data.to_csv('./data/4.1/data1.csv',header=True,index=False)
```
