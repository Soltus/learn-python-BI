```python
import pandas as pd
import statsmodels.api as sm #最小二乘
from statsmodels.formula.api import ols #加载ols模型
```

```python
data = pd.read_csv('./data/4.1/data2.csv')
```

```python
data
```

```python
data.rename(columns={'monthly_sales_log':'月销量', 'price_log':'价格'}, inplace = True)
```

```python
data
```

```python
lm=ols('月销量~价格+圆领+韩版+黑色+中长款+白色+显瘦+雪纺+条纹+荷叶边+蓝色+红色+粉色',data = data).fit()
```

```python
print(lm.summary())
```

```python
lm=ols('月销量~价格+韩版+中长款+白色+显瘦+粉色',data = data).fit()
```

```python
print(lm.summary())
```

```python
lm=ols('月销量~价格+韩版+中长款+白色+显瘦+粉色+age17*韩版+age18_24*韩版+
       age25_29*韩版+age30_34*韩版+age35_39*韩版+age40*韩版',data = data).fit()
print(lm.summary())
```
