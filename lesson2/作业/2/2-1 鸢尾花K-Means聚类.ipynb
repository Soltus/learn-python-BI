```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
iris = load_iris()  #由于iris数据内置sklearn，调用load_iris()方法便可获取数据
iris_data = iris['data'] ##提取数据集中的特征
iris_target = iris['target'] ## 提取数据集中的标签
iris_names = iris['feature_names'] ### 提取特征名
```

```python
iris_data
```

```python
iris_target
```

```python
iris_names #花萼长度、花萼宽度、花瓣长度、花瓣宽度
```

```python
scale = MinMaxScaler().fit(iris_data)## 训练规则
iris_dataScale = scale.transform(iris_data) ## 应用规则
iris_dataScale
```

```python
kmeans = KMeans(n_clusters = 3,random_state=123).fit(iris_dataScale) ##构建并训练模型
print('构建的K-Means模型为：\n',kmeans)
```

```python
kmeans.labels_
```

```python
import pandas as pd
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
##使用TSNE进行数据降维,降成两维
tsne = TSNE(n_components=2,init='random',random_state=177).fit(iris_data)
df=pd.DataFrame(tsne.embedding_) ##将原始数据转换为DataFrame
df['labels'] = kmeans.labels_ ##将聚类结果存储进df数据表
```

```python
tsne.embedding_
```

```python
df
```

```python
##提取不同标签的数据
df1 = df[df['labels']==0]
df2 = df[df['labels']==1] 
df3 = df[df['labels']==2] 
```

```python
## 绘制图形
fig = plt.figure(figsize=(6,4)) ##设定空白画布，并制定大小
##用不同的颜色表示不同数据
plt.plot(df1[0],df1[1],'bo',df2[0],df2[1],'r*',df3[0],df3[1],'gD')
plt.show() ##显示图片
```

```python
import numpy as np
a =  [1.5, 1.5, 1.5, 1.5]      

result = kmeans.predict([[(a[0]-np.min(iris_data,axis=0)[0])/(np.max(iris_data,axis=0)[0]-np.min(iris_data,axis=0)[0]),
                          (a[1]-np.min(iris_data,axis=0)[1])/(np.max(iris_data,axis=0)[1]-np.min(iris_data,axis=0)[1]),
                          (a[2]-np.min(iris_data,axis=0)[2])/(np.max(iris_data,axis=0)[2]-np.min(iris_data,axis=0)[2]),
                          (a[3]-np.min(iris_data,axis=0)[3])/(np.max(iris_data,axis=0)[3]-np.min(iris_data,axis=0)[3])]])
print('花瓣花萼长度宽度全为1.5的鸢尾花预测类别为：', result[0])
```

```python
from sklearn.metrics import fowlkes_mallows_score
for i in range(2,7):
    ##构建并训练模型
    kmeans = KMeans(n_clusters = i,random_state=123).fit(iris_dataScale)
    score = fowlkes_mallows_score(iris_target,kmeans.labels_)
    print('iris数据聚%d类FMI评价分值为：%f' %(i,score))
```
